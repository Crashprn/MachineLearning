```python
import numpy as np
import matplotlib.pyplot as plt
```

# Defining X and Y

```python
x = np.linspace(-10, 10, 1000)
y = np.linspace(-10, 10, 1000)

X, Y = np.meshgrid(x, y)
```

## Part 1

```python
plt.contourf(X, Y, np.cos(X-Y))
plt.title('Contour plot of cos(x-y)')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar()
plt.show()
```

# Part 2

```python
plt.contourf(X, Y, np.cos(X**2 - Y**2))
plt.title('Contour plot of cos(x^2 - y^2)')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Part 3

```python
plt.contourf(X, Y, np.tanh(X*Y + 1))
plt.title('Contour plot of tanh(xy + 1)')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
